```python
#Task3
n = int (input("Enter number of elements: "))
list1 = []
count = 0
odd_count=0
for i in range(n):
    num = int (input(f"Enter number {i+1}: "))
    list1.append(num)
    print("List: ", list1)
    if num%2 == 0:
        count += 1
        
    else:
        odd_count+=1
       
print("Even count is ", count)
print("Odd count is", odd_count)

index = int (input("Enter the index of the value that you want to replace"))
new_value = int (input("Enter the new value"))
if 0 <= index < len(list1):
    list1[index] = new_value
    print("Updated list: ", list1)
else:
    print("Invalid Index")

```

```python
#Task 4
def total_salary(salary):
    HRA = salary * 0.1
    DA = salary * 0.05
    total = HRA + DA + salary
    return total 

salary = float (input("Enter your salary"))
result = total_salary(salary)
print("Total salary: ", result)
```

```python
#Task 5
def string_analysis(string1):
    dict1={
        "total characters" : len(string1),
        "uppercase" : 0,
        "lowercase" : 0,
        "no. of digits" : 0
    }

    for ch in string1:
        if ch.isupper():
            dict1["uppercase"] += 1
        elif ch.islower():
            dict1["lowercase"] += 1
        elif ch.isdigit():
            dict1["no. of digits"] += 1
    return dict1
user_input = input("Enter a string")
output = string_analysis(user_input)
output
        
```

```python
#Task 6
s = input("Enter a string: ")

longest = ""
current = ""

for ch in s:
    if current == "" or ch == current[-1]:
        current += ch
    else:
        if len(current) > len(longest):
            longest = current
        current = ch


if len(current) > len(longest):
    longest = current

print("Longest substring:", longest)
print("Length:", len(longest))



    
```

```python
#Task 7
def find_repeated_numbers(numbers):
    repeated = []

    for i in range(len(numbers)):
        for j in range(i + 1, len(numbers)):
            if numbers[i] == numbers[j]:
                if numbers[i] not in repeated:
                    repeated.append(numbers[i])

    return repeated



n = int(input("Enter number of elements: "))
lst = []

for i in range(n):
    lst.append(int(input(f"Enter number {i+1}: ")))


result = find_repeated_numbers(lst)

print("Repeated numbers:", result)
```
