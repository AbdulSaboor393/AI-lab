```python
class SecurityAgent:
    def __init__(self, agent_id, name, status):
        self.agent_id = agent_id
        self.name = name
        self.status = status

    def show_status(self):
        print(self.name, "is currently", self.status)


class FirewallAgent(SecurityAgent):
    def monitor_traffic(self):
        print(self.name, "is monitoring network traffic and blocking suspicious activity")


class MalwareDetectionAgent(SecurityAgent):
    def scan_files(self):
        print(self.name, "is scanning files and removing malware threats")


class AutomationAgent(SecurityAgent):
    def run_automation(self):
        print(self.name, "is running automated tasks to improve system security")


firewall = FirewallAgent(1, "Firewall Agent", "Active")
malware = MalwareDetectionAgent(2, "Malware Detection Agent", "Active")
automation = AutomationAgent(3, "Automation Agent", "Active")

firewall.show_status()
firewall.monitor_traffic()

malware.show_status()
malware.scan_files()

automation.show_status()
automation.run_automation()
```

```python
class Threat:
    def __init__(self, threat_id, name, severity):
        self.threat_id = threat_id
        self.name = name
        self.severity = severity

    def show_details(self):
        print(self.name, "threat detected with severity level", self.severity)


class PhishingThreat(Threat):
    def analyze_email(self):
        print(self.name, "is analyzing emails and blocking phishing attempts")


class RansomwareThreat(Threat):
    def scan_files(self):
        print(self.name, "is scanning files and preventing ransomware attacks")


class BotnetThreat(Threat):
    def detect_traffic(self):
        print(self.name, "is monitoring network traffic for botnet behavior")


phishing = PhishingThreat(101, "Phishing", "High")
ransomware = RansomwareThreat(102, "Ransomware", "Critical")
botnet = BotnetThreat(103, "Botnet", "Medium")

phishing.show_details()
phishing.analyze_email()

ransomware.show_details()
ransomware.scan_files()

botnet.show_details()
botnet.detect_traffic()
```

```python
class ResponseAgent:
    def execute_response(self):
        print("Executing a general response")


class AlertAgent(ResponseAgent):
    def execute_response(self):
        print("Alert sent to the security team")


class BlockAgent(ResponseAgent):
    def execute_response(self):
        print("Malicious activity has been blocked")


class RecoverAgent(ResponseAgent):
    def execute_response(self):
        print("System recovery process has started")


agents = [
    AlertAgent(),
    BlockAgent(),
    RecoverAgent()
]

for agent in agents:
    agent.execute_response()
```

```python
class Employee:
    def work(self):
        print("Employee is working")


class Manager(Employee):
    def work(self):
        print("Manager is planning and supervising the team")


class Developer(Employee):
    def work(self):
        print("Developer is writing and debugging code")


class Designer(Employee):
    def work(self):
        print("Designer is creating visual designs")


employees = [
    Manager(),
    Developer(),
    Designer()
]

for emp in employees:
    emp.work()
```

```python
class Student:
    def __init__(self, name):
        self.name = name
        self.__grade = None

    def set_grade(self, grade):
        self.__grade = grade

    def get_grade(self):
        return self.__grade

    def display_info(self):
        print("Student Name:", self.name)
        print("Grade:", self.__grade)


student1 = Student("Ali")
student2 = Student("Sara")

student1.set_grade(85)
student2.set_grade(92)

print(student1.get_grade())
print(student2.get_grade())

student1.display_info()
student2.display_info()
```
