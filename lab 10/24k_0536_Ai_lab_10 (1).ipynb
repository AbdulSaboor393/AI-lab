```python
import numpy as np

intel = {"High": 0.7, "Low": 0.3}
study = {"Sufficient": 0.6, "Insufficient": 0.4}
difficulty = {"Hard": 0.4, "Easy": 0.6}

grades = {
    ("High", "Sufficient", "Easy"):   {"A": 0.7, "B": 0.2, "C": 0.1},
    ("High", "Sufficient", "Hard"):   {"A": 0.5, "B": 0.3, "C": 0.2},
    ("High", "Insufficient", "Easy"): {"A": 0.4, "B": 0.4, "C": 0.2},
    ("High", "Insufficient", "Hard"): {"A": 0.2, "B": 0.4, "C": 0.4},
    ("Low", "Sufficient", "Easy"):    {"A": 0.3, "B": 0.4, "C": 0.3},
    ("Low", "Sufficient", "Hard"):    {"A": 0.1, "B": 0.3, "C": 0.6},
    ("Low", "Insufficient", "Easy"):  {"A": 0.1, "B": 0.3, "C": 0.6},
    ("Low", "Insufficient", "Hard"):  {"A": 0.05, "B": 0.2, "C": 0.75}
}

pass_chance = {"A": 0.95, "B": 0.80, "C": 0.50}

def calc_joint(i, s, d, g):
    return intel[i] * study[s] * difficulty[d] * grades[(i, s, d)][g]

intel_vals = ["High", "Low"]
study_vals = ["Sufficient", "Insufficient"]
diff_vals = ["Hard", "Easy"]
grade_vals = ["A", "B", "C"]

print("\nQuery 1")

den = 0
num = 0

for i in intel_vals:
    for g in grade_vals:
        val = calc_joint(i, "Sufficient", "Hard", g)
        den += val
        num += val * pass_chance[g]

ans1 = num / den
print(f"Result = {ans1:.4f}")

print("\nQuery 2")

intel_pass = {"High": 0, "Low": 0}
total_pass = 0

for i in intel_vals:
    for s in study_vals:
        for d in diff_vals:
            for g in grade_vals:
                val = calc_joint(i, s, d, g)
                p = val * pass_chance[g]
                intel_pass[i] += p
                total_pass += p

ans2 = intel_pass["High"] / total_pass
print(f"Result = {ans2:.4f}")
```

```python
prior = {"Flu": 0.3, "Cold": 0.7}

fever_prob   = {"Flu": {"Yes": 0.9, "No": 0.1}, "Cold": {"Yes": 0.5, "No": 0.5}}
cough_prob   = {"Flu": {"Yes": 0.8, "No": 0.2}, "Cold": {"Yes": 0.6, "No": 0.4}}
fatigue_prob = {"Flu": {"Yes": 0.7, "No": 0.3}, "Cold": {"Yes": 0.3, "No": 0.7}}
chills_prob  = {"Flu": {"Yes": 0.6, "No": 0.4}, "Cold": {"Yes": 0.4, "No": 0.6}}

def calculate_prob(obs):
    values = {}
    for d in ["Flu", "Cold"]:
        val = prior[d]

        if "Fever" in obs:
            val *= fever_prob[d][obs["Fever"]]
        if "Cough" in obs:
            val *= cough_prob[d][obs["Cough"]]
        if "Fatigue" in obs:
            val *= fatigue_prob[d][obs["Fatigue"]]
        if "Chills" in obs:
            val *= chills_prob[d][obs["Chills"]]

        values[d] = val

    s = sum(values.values())
    return {k: values[k] / s for k in values}

print("\nCase 1: Fever=Yes, Cough=Yes")
res1 = calculate_prob({"Fever": "Yes", "Cough": "Yes"})
for k, v in res1.items():
    print(f"  {k}: {v:.4f}")

print("\nCase 2: Fever=Yes, Cough=Yes, Chills=Yes")
res2 = calculate_prob({"Fever": "Yes", "Cough": "Yes", "Chills": "Yes"})
for k, v in res2.items():
    print(f"  {k}: {v:.4f}")

print("\nCase 3:")
print(f"  P(Fatigue=Yes | Flu) = {fatigue_prob['Flu']['Yes']:.4f}")
```

```python
import numpy as np

weather_states = ["Sunny", "Cloudy", "Rainy"]

transitions = np.array([
    [0.7, 0.2, 0.1],
    [0.3, 0.4, 0.3],
    [0.2, 0.3, 0.5]
])

def simulate_weather(n_days):
    sequence = ["Sunny"]
    for _ in range(n_days - 1):
        current = sequence[-1]
        position = weather_states.index(current)
        next_state = np.random.choice(weather_states, p=transitions[position])
        sequence.append(next_state)
    return sequence

experiments = []
for _ in range(10000):
    seq = simulate_weather(10)
    experiments.append(seq.count("Rainy") >= 3)

probability = sum(experiments) / len(experiments)

print(f"Sequence: {' -> '.join(simulate_weather(10))}")
print(f"Probability of >= 3 rainy days: {probability}")
```
