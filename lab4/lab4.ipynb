```python

class goalbasedagent:
    def __init__(self, graph, goal, depth_limit):
        self.graph = graph
        self.goal = goal
        self.depth_limit = depth_limit

    def dls(self, current, depth):
        print("visiting:", current, "depth:", depth)

        if current == self.goal:
            print("goal found:", current)
            return True

        if depth == self.depth_limit:
            return False

        for neighbor in self.graph[current]:
            if self.dls(neighbor, depth + 1):
                return True

        return False


graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

agent1 = goalbasedagent(graph, 'E', 2)
agent1.dls('A', 0)
```

```python

class utilitybasedagent:
    def __init__(self, graph, goal):
        self.graph = graph
        self.goal = goal

    def ucs(self, start):
        queue = []
        queue.append((0, start))

        visited = []

        while len(queue) > 0:

            queue.sort()

            cost, node = queue.pop(0)

            if node not in visited:
                print("visiting:", node, "cost:", cost)
                visited.append(node)

                if node == self.goal:
                    print("goal found with cost:", cost)
                    return

                for neighbor, weight in self.graph[node]:
                    queue.append((cost + weight, neighbor))


graph2 = {
    'A': [('B', 1), ('C', 4)],
    'B': [('D', 2), ('E', 5)],
    'C': [('F', 1)],
    'D': [],
    'E': [],
    'F': []
}

agent2 = utilitybasedagent(graph2, 'F')
agent2.ucs('A')
```

```python
graph = {
    1: [(2,10), (3,15), (4,20)],
    2: [(1,10), (3,35), (4,25)],
    3: [(1,15), (2,35), (4,30)],
    4: [(1,20), (2,25), (3,30)]
}

start = 1
total_cities = 4

queue = []
queue.append((0, start, [start], [start]))

min_cost = 999999
best_path = []

while len(queue) > 0:

    queue.sort()
    cost, current, visited, path = queue.pop(0)

    if len(visited) == total_cities and current == start:
        if cost < min_cost:
            min_cost = cost
            best_path = path
        continue

    for neighbor, weight in graph[current]:

        if neighbor not in visited:
            new_visited = visited + [neighbor]
            new_path = path + [neighbor]
            queue.append((cost + weight, neighbor, new_visited, new_path))

        if len(visited) == total_cities and neighbor == start:
            new_path = path + [start]
            queue.append((cost + weight, start, visited, new_path))

print("shortest path:", best_path)
print("minimum cost:", min_cost)
```

```python
class iddfs:
    def __init__(self, structure, is_graph=False):
        self.s = structure
        self.g = is_graph

    def dfs(self, node, goal, depth, path, vis=None):
        path.append(node)
        if node == goal:
            return path
        if depth <= 0:
            path.pop()
            return None
        for n in self.s.get(node, []):
            if self.g:
                if n not in vis:
                    vis.add(n)
                    r = self.dfs(n, goal, depth-1, path, vis)
                    if r: return r
                    vis.remove(n)
            else:
                r = self.dfs(n, goal, depth-1, path)
                if r: return r
        path.pop()
        return None

    def search(self, start, goal, maxd=50):
        for d in range(maxd):
            path = []
            vis = set([start]) if self.g else None
            r = self.dfs(start, goal, d, path, vis)
            if r: return r
        return None

tree = {'a':['b','c'],'b':['d','e'],'c':['f'],'d':[],'e':[],'f':[]}
graph = {'a':['b','c'],'b':['a','d','e'],'c':['a','f'],'d':['b'],'e':['b'],'f':['c']}

t = iddfs(tree)
print("tree path:", t.search('a','e'))

g = iddfs(graph, True)
print("graph path:", g.search('a','e'))
```
