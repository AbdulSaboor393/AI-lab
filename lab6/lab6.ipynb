```python

class GoalAgent:
    def __init__(self, goal, depth_limit):
        self.goal = goal
        self.depth_limit = depth_limit

    def check_goal(self, current):
        if current == self.goal:
            return "Goal Reached"
        return "Searching"

    def depth_limited_search(self, graph, start, goal, depth_limit):
        explored = []

        def explore(node, level):
            if level > depth_limit:
                return None
            explored.append(node)
            if node == goal:
                print(f"Goal found with DLS. Path: {explored}")
                return explored
            for neighbor in graph.get(node, []):
                if neighbor not in explored:
                    path = explore(neighbor, level + 1)
                    if path:
                        return path
            explored.pop()
            return None

        return explore(start, 0)

    def action(self, current, graph):
        goal_state = self.check_goal(current)
        if goal_state == "Goal Reached":
            return f"Goal {self.goal} found!"
        else:
            return self.depth_limited_search(graph, current, self.goal, self.depth_limit)


graph_structure = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': ['H'],
    'E': [],
    'F': ['I'],
    'G': [],
    'H': [],
    'I': []
}
start_node = 'A'
goal_node = 'I'
depth_limit_value = 3

agent_instance = GoalAgent(goal_node, depth_limit_value)
print(agent_instance.action(start_node, graph_structure))
```

```python
# Task #1: Uniform Cost Search (UCS) as a Utility-Based Agent for Raahim

import heapq

class UtilityAgent:
    def __init__(self, goal):
        self.goal = goal

    def search_ucs(self, graph, start, goal):
        frontier = [(0, start)]
        visited = set()
        cost_tracker = {start: 0}
        path_tracker = {start: None}

        while frontier:
            current_cost, current_node = heapq.heappop(frontier)

            if current_node in visited:
                continue
            visited.add(current_node)

            if current_node == goal:
                path = []
                while current_node is not None:
                    path.append(current_node)
                    current_node = path_tracker[current_node]
                path.reverse()
                print(f"Goal found with UCS. Path: {path}, Total Cost: {current_cost}")
                return path

            for neighbor, cost in graph[current_node].items():
                new_cost = current_cost + cost
                if neighbor not in cost_tracker or new_cost < cost_tracker[neighbor]:
                    cost_tracker[neighbor] = new_cost
                    path_tracker[neighbor] = current_node
                    heapq.heappush(frontier, (new_cost, neighbor))

        print("Goal not found")
        return None

    def perform_action(self, graph, start, goal):
        return self.search_ucs(graph, start, goal)


graph_structure = {
    'A': {'B': 2, 'C': 1},
    'B': {'D': 4, 'E': 3},
    'C': {'F': 1, 'G': 5},
    'D': {'H': 2},
    'E': {},
    'F': {'I': 6},
    'G': {},
    'H': {},
    'I': {}
}

start_node = 'A'
goal_node = 'I'

agent_instance = UtilityAgent(goal_node)
print(agent_instance.perform_action(graph_structure, start_node, goal_node))
```

```python

import itertools

cost_matrix = {
    'A': {'B': 10, 'C': 15, 'D': 20},
    'B': {'A': 10, 'C': 35, 'D': 25},
    'C': {'A': 15, 'B': 35, 'D': 30},
    'D': {'A': 20, 'B': 25, 'C': 30}
}

def calculate_cost(route):
    total_cost = 0
    for i in range(len(route) - 1):
        total_cost += cost_matrix[route[i]][route[i + 1]]
    total_cost += cost_matrix[route[-1]][route[0]]
    return total_cost

cities = list(cost_matrix.keys())
possible_routes = itertools.permutations(cities)

shortest_route = None
min_cost = float('inf')

for route in possible_routes:
    route_cost = calculate_cost(route)
    if route_cost < min_cost:
        shortest_route = route
        min_cost = route_cost

print(f"The shortest route is {shortest_route} with a total cost of {min_cost}")
```
