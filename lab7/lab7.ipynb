```python
graph = {
    'S': [('A', 2), ('B', 3)],
    'A': [('C', 4)],
    'B': [('D', 5)],
    'C': [('E', 6)],
    'D': [], 'E': []
}

def search(start, goal, k):
    beam = [(start, 0, [start])]

    while beam:
        next_level = []

        for node, cost, path in beam:
            if node == goal:
                return path, cost

            for n, w in graph[node]:
                next_level.append((n, cost + w, path + [n]))

        next_level.sort(key=lambda x: x[1])
        beam = next_level[:k]

    return None, -1

print(search('S', 'E', 2))
```

```python
import random

def d(a,b):
    return abs(a[0]-b[0]) + abs(a[1]-b[1])

def calc(r):
    s = 0
    for i in range(len(r)-1):
        s += d(r[i], r[i+1])
    return s

def solve(p):
    cur = p[:]
    random.shuffle(cur)

    while True:
        improved = False
        for i in range(len(cur)):
            for j in range(i+1, len(cur)):
                new = cur[:]
                new[i], new[j] = new[j], new[i]

                if calc(new) < calc(cur):
                    cur = new
                    improved = True

        if not improved:
            break

    return cur, calc(cur)

pts = [(1,1),(2,5),(3,2),(7,4)]
print(solve(pts))
```

```python
import random

cities = [(1,1),(2,2),(3,3),(4,4)]

def f(r):
    d = 0
    for i in range(len(r)-1):
        d += abs(r[i][0]-r[i+1][0]) + abs(r[i][1]-r[i+1][1])
    return d

def gen():
    r = cities[:]
    random.shuffle(r)
    return r

def mix(a,b):
    c = []
    for x in a:
        if x not in c:
            c.append(x)
    for x in b:
        if x not in c:
            c.append(x)
    return c

def change(r):
    i,j = random.sample(range(len(r)),2)
    r[i],r[j] = r[j],r[i]
    return r

pop = [gen() for _ in range(5)]

for _ in range(30):
    pop.sort(key=f)
    best = pop[:2]

    new = best[:]
    while len(new)<5:
        a,b = random.sample(best,2)
        c = mix(a,b)
        if random.random()<0.1:
            c = change(c)
        new.append(c)

    pop = new

print(min(pop,key=f))
```

```python
tasks = [2,4,6]
p = [0,0]

def run(k):
    beam = [p[:]]

    for t in tasks:
        temp = []
        for b in beam:
            for i in range(len(b)):
                n = b[:]
                n[i] += t
                temp.append(n)

        temp.sort(key=lambda x: max(x))
        beam = temp[:k]

    return beam[0]

print(run(2))
```
